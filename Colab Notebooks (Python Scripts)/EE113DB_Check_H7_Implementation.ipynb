```
#import libraries for nerualnet and math
from __future__ import absolute_import, division, print_function, unicode_literals

try:
  # %tensorflow_version only exists in Colab.
  %tensorflow_version 2.x
except Exception:
  pass
import tensorflow as tf
import numpy as np
from tensorflow.keras import datasets, layers, models
print(tf.__version__)
```

```
#part a) build a verification model for convolutional layer
conv_ck = tf.keras.Sequential(tf.keras.layers.Conv2D(2, (3, 3), activation='relu', input_shape=(5, 5, 1), padding='same'))
conv_ck.summary()
```

```
#create a numpy array to prepare parameter to be set for model
kernal_1 = np.asarray([[[[1, 3]], [[2, 3]], [[3, 3]]], [[[4, 3]], [[5, 3]], [[6, 3]]], [[[7, 3]], [[8, 3]], [[9, 3]]]])
print(kernal_1)
bias_1 = np.asarray([0.5, 1])
#set models parameter as your choice, so that LCDK implementation can be checked
conv_ck.layers[0].set_weights([kernal_1, bias_1])
```

```
#print out the current parameter for a double check
print(conv_ck.layers[0].get_weights())
```

```
#create a input data to check for result
input_tmp = np.ones(25).reshape([1, 5, 5, 1])
print("input matrix is\n", input_tmp)
print("-------------------")
print("output matrix is \n", conv_ck.predict(input_tmp))
```

```
t1, t2 = conv_ck.layers[0].get_weights()
np.savetxt('wtest.txt', t1.flatten(), delimiter=',',fmt='%.16f') 
np.savetxt('btest.txt', t2.flatten(), delimiter=',',fmt='%.16f') 
np.savetxt('input.txt', input_tmp.flatten(), delimiter=',',fmt='%.16f')
np.savetxt('expectedresult.txt', conv_ck.predict(input_tmp).flatten(), delimiter=',',fmt='%.16f')
```

```
#a more complicated one to check for multichannel input, multichannel output
conv_ck_2 = tf.keras.Sequential(tf.keras.layers.Conv2D(3, (3, 3), activation='relu', input_shape=(5, 5, 2), padding='same'))
conv_ck_2.summary()
```

```
#create a numpy array to prepare parameter to be set for model
kernal_2 = np.asarray([[[[1, 3, 5], [1, 2, 0]], [[2, 3, 5], [1, 2, 0]], [[3, 3, 5], [1, 2, 0]]], [[[4, 3, 5], [1, 2, 0]], [[5, 3, 5], [1, 2, 0]], [[6, 3, 5], [1, 2, 0]]], [[[7, 3, 5], [1, 2, 0]], [[8, 3, 5], [1, 2, 0]], [[9, 3, 5], [1, 2, 0]]]])
bias_2 = np.asarray([0.5, 1, 2.5])
#set models parameter as your choice, so that LCDK implementation can be checked
conv_ck_2.layers[0].set_weights([kernal_2, bias_2])
```

```
#create a input data to check for result
input_tmp_2 = np.tile(np.asarray([[1, 2.5], [1, 2.5], [1, 2.5], [1, 2.5], [1, 2.5]]), (5, 1, 1)).reshape([1, 5, 5, 2])
print("input matrix is\n", input_tmp_2)
print("-------------------")
print("output matrix is \n", conv_ck_2.predict(input_tmp_2))
```

```
t1, t2 = conv_ck_2.layers[0].get_weights()
np.savetxt('wtest2.txt', t1.flatten(), delimiter=',',fmt='%.16f') 
np.savetxt('btest2.txt', t2.flatten(), delimiter=',',fmt='%.16f') 
np.savetxt('input2.txt', input_tmp_2.flatten(), delimiter=',',fmt='%.16f')
np.savetxt('res2.txt', conv_ck_2.predict(input_tmp_2).flatten(), delimiter=',',fmt='%.16f')
```

```
#part b) build a verification model for pooling layer
pooling_ck = tf.keras.Sequential(tf.keras.layers.MaxPooling2D((2, 2), input_shape=(4, 4, 2)))
pooling_ck.summary()
```

```
input_tmp = np.reshape(np.asarray([[[1, 4.5], [2, 3.5], [3, 2.5], [4, 1.5]], [[5, 8.5], [6, 7.5], [7, 6.5], [8, 5.5]], [[9, 12.5], [10, 11.5], [11, 10.5], [12, 9.5]], [[13, 16.5], [14, 15.5], [15, 14.5], [16, 13.5]]]), [1, 4, 4, 2])
print("input matrix is\n", input_tmp)
print("-------------------")
print("output matrix is \n", pooling_ck.predict(input_tmp))
```

```
np.savetxt('input3.txt', input_tmp.flatten(), delimiter=',',fmt='%.16f')
np.savetxt('res3.txt', pooling_ck.predict(input_tmp).flatten(), delimiter=',',fmt='%.16f')
```

```
#part c) build a verification model for dense layer
#use last prediction result just to save some effort
t1 = np.asarray([[[2, 4.5], [2, 3]], [[3, 3], [2, 2.2]]]).reshape([1, 2, 2, 2])
dense_ck = models.Sequential()
dense_ck.add(layers.Flatten(input_shape=(2, 2, 2)))
dense_ck.add(layers.Dense(10, activation='softmax'))
dense_ck.summary()
```

```
kernel_3 = np.asarray([np.arange(x+1, x+6, 0.5) for x in range(8)])
bias_3 = np.arange(0, 10, 1)
```

```
dense_ck.layers[1].set_weights([kernel_3, bias_3])
dense_ck.layers[1].get_weights()
```

```
dense_ck.predict(t1)
```

```
np.savetxt('wtest4.txt', kernel_3.flatten(), delimiter=',',fmt='%.16f') 
np.savetxt('btest4.txt', bias_3.flatten(), delimiter=',',fmt='%.16f') 
np.savetxt('input4.txt', t1.flatten(), delimiter=',',fmt='%.16f')
np.savetxt('res4.txt', dense_ck.predict(t1).flatten(), delimiter=',',fmt='%.16f')
```
